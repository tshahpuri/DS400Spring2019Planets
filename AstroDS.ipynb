## Task 1 Baseline model

Creating a baseline model with gamma(30.87, 0.271) and Expected_MES attribute to predict probability of attribute Recovered_Flag


```
prob = []
for i in range(len(data["Expected_MES"])):
    a,b = (0.94 * gamma.cdf((30.87,0.271), (data["Expected_MES"][i])))
    prob.append(b)
```

Using Root Mean Square Error as our 

```
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```

Checking our error

```
rmse(prob, data["Recovered"])
```
